```python
#import dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Extract Mass Shootings CSV into Dataframes

mass_shootings_all="Resources/mass_shootings_all.csv"
mass_shootings_all_df=pd.read_csv(mass_shootings_all)
mass_shootings_all_df.head()

```

```python
#Extract Accidental Deaths CSV into Dataframes

accidental_deaths="Resources/accidental_deaths.csv"
accidental_deaths_df=pd.read_csv(accidental_deaths)
accidental_deaths_df.head()
```

```python
#Extract Officer Involved Shootings CSV into Dataframes

officer_involved_shootings="Resources/officer_involved_shootings.csv"
officer_involved_shootings_df=pd.read_csv(officer_involved_shootings)
officer_involved_shootings_df.head()
```

```python
#Transform and clean Mass Shootings DataFrames
mass_shootings_cols = ["Incident Date", "State", "# Killed", "# Injured"]
mass_shootings_transformed = mass_shootings_all_df[mass_shootings_cols].copy()

mass_shootings_transformed = mass_shootings_transformed.rename(columns={"Incident Date": "incident_date",
                                                                       "State": "state",
                                                                       "# Killed": "killed",
                                                                       "# Injured": "injured"})

mass_shootings_transformed.head()
```

```python
#Transform and clean Accidental DataFrames
accidental_deaths_cols = ["Incident Date", "State", "# Killed", "# Injured"]
accidental_deaths_transformed = accidental_deaths_df[accidental_deaths_cols].copy()

accidental_deaths_transformed = accidental_deaths_transformed.rename(columns={"Incident Date": "incident_date",
                                                                       "State": "state",
                                                                       "# Killed": "killed",
                                                                       "# Injured": "injured"})

accidental_deaths_transformed.head()
```

```python
#Transform and clean Officer involved DataFrames
officer_involved_shootings_cols = ["Incident Date", "State", "# Killed", "# Injured"]
officer_involved_shootings_transformed = officer_involved_shootings_df[officer_involved_shootings_cols].copy()

officer_involved_shootings_transformed = officer_involved_shootings_transformed.rename(columns={"Incident Date": "incident_date",
                                                                       "State": "state",
                                                                       "# Killed": "killed",
                                                                       "# Injured": "injured"})

officer_involved_shootings_transformed.head()
```

```python
#Create Database Connection
connection_string = "postgres:postgres@localhost:5432/gun_violence_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
#Verify tables in Database
engine.table_names()
```

```python
#Load Mass Shooting DF into Database
mass_shootings_transformed.to_sql(name='mass_shootings_all', con=engine, if_exists='append', index=True)
```

```python
#Load Officer Involved Shootings Shooting DF into Database
officer_involved_shootings_transformed.to_sql(name='officer_involved_shootings', con=engine, if_exists='append', index=True)
```

```python
#Load Accident death  DF into Database
accidental_deaths_transformed.to_sql(name='accidental_deaths', con=engine, if_exists='append', index=True)
```
